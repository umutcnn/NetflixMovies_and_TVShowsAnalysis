```python
sns.countplot(data["type"])
fig=plt.gcf()
fig.set_size_inches(5,5)
plt.title("tur")
```

```python
sns.countplot(data["rating"])
sns.countplot(data["rating"]).set_xticklabels(sns.countplot(data["rating"]).get_xticklabels(),rotation=90,ha="right")

fig=plt.gcf()
fig.set_size_inches(13,13)
plt.title("rating")
```

```python
#film ve tv programlarının karşılaştırılması
plt.figure(figsize=(10,8))
sns.countplot(x="rating", hue="type",data=data)
plt.title("tür ve raiting oranları")
plt.show()
```

```python
#film ve tv programlarının karşılaştırılması pasta grafiği
labels = ["Movie", "TV Show"]
size = data["type"].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]  # Adjusted to match the length of 'size'

plt.rcParams["figure.figsize"] = (9, 9)

plt.pie(size, labels=labels, colors=colors, explode=explode, shadow=True, startangle=90)
plt.title("Tür Görünümü", fontsize=25)
plt.legend()
plt.show()
```

```python
#farklı film türleri, puanlar veya diğer değerlendirme kategorileri
plt.figure(figsize=(10, 8))

data["rating"].value_counts().plot.pie(autopct="%1.1f%%", shadow=True)

plt.show()
```
